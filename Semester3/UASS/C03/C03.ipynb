```python
# 1. Doplňtě kód, použíjte pouze metody v tomto souboru
# 2. Vypočíteje pro náhodnou síť a síť s pref. připojováním z minula:
  # velikost sítě
  # průměrný a min / max stupeň
  # hustotu sítě
  # průměrný shlukovací koeficient

# 3. Vypočítejte zároveň vlastnosti i pro jednu z vybraných sítí:

    # Dophins
    # LesMiserables
    # (V souborech jsou dve verze, jedna pro Gephi a druha CSV pro tento kod)

# 4. Doplnte, do reportu z minula, výsledné vlastnosti pro každou síť, do přehledné tabulky.
# 5. Porovnejte s výsledky z Gephi.

import networkx as nx
from statistics import mean


# TODO - zmente nazev site, popripadne delimiter
#G = nx.read_edgelist("C02_real.csv", delimiter=';',nodetype=str) # TODO - doplnit cestu k souboru, pripadne zmenit delimiter
#G = nx.read_edgelist("Dolphins.csv", delimiter=';',nodetype=str)
G = nx.read_edgelist("C02_unreal.csv", delimiter=';',nodetype=str)

# Vrati seznam vrcholu
def nodes() -> list[str]:
    return list(G.nodes())


# Seznam hran
def edges() -> list[tuple[str,str]]:
    return list(G.edges())


# Hustota
def net_density():
    density = len(edges()) / ((len(nodes()) * (len(nodes()) - 1)) / 2)
    # TODO - vypoctete hustotu site, podle vzorce na webu (metoda "len" vrati velikost seznamu, napr. len(edges)), s pouzitim metod nodes a edges

    return density


# Stupen vrcholu node
def local_degree(node:str) -> int:
    degrees = {}

    if node not in degrees:
                             # https://networkx.org/documentation/stable/reference/classes/generated/networkx.Graph.degree.html
        degrees[node] = G.degree(node) # TODO - Nahradte None stupnem vrcholu "node", pouzijte metodu z networkx, viz. vyse.


    return degrees[node]


# Seznam stupnu pro vsechny vrcholy
def nodes_degrees() -> list[int]:
    return [local_degree(i) for i in nodes()]


# Prumerny stupen
def average_degree():

    # TODO - doplnte vypocet prumerneho stupne (metoda mean), pouzijte metodu nodes_degrees
    average_degree = mean(nodes_degrees())



    return average_degree


# Nejmensi stupen v siti
def min_degree():

    # TODO - doplnte vypocet min. stupne (metoda min), pouzijte metodu nodes_degrees
    min_degree = min(nodes_degrees())



    return min_degree


# Nejvetsi stupen v siti
def max_degree():

    # TODO - doplnte vypocet max. stupne (metoda max), pouzijte metodu nodes_degrees
    max_degree = max(nodes_degrees())


    return max_degree

# Vypocet lokalniho shlukovaciho koeficientu pro vrchol node
#
def local_clustering_coefficient(node: str) -> float:
    # Okoli vrcholu node, pro urceni poctu hran v okoli node
    G_induced = nx.Graph(nx.induced_subgraph(G, nx.neighbors(G, node)))

    L_i = len(G_induced.edges)
    k_i = local_degree(node)

    if k_i > 1:
        # TODO - doplnit vzorec pro lokalni shlukovaci koeficient - na webu
        return (2 * L_i) / (k_i * (k_i - 1))
    else:
        return 0

# Prumerny shlukovaci koeficient
def average_clustering_coefficient():
    return mean([local_clustering_coefficient(i) for i in nodes()])


# print(G)
print(f"Sit ma {len(nodes())} vrcholu a {len(edges())} hran.")
print(f"Prumerny stupen je {average_degree()}, min. stupen je {min_degree()} a max. stupen je {max_degree()}")
print(f"Hustota site je {net_density()}")
print(f"Prumerny shlukovaci koeficient je {average_clustering_coefficient()}")
```
