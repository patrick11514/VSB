```r
data <- readxl::read_excel("FPS_data.xlsx", sheet = "Vysledky mereni", skip = 0)
head(data)
```

```r
df_long <- data %>%
  pivot_longer(
    cols = -id,  # Vezme všechny sloupce KROMĚ 'id'
    names_to = c("Karta", "Status"), # Do jakých nových sloupců chceme jména
    # Tento regulární výraz rozdělí jména sloupců
    # (.*) = 1. část (Karta)
    # (.*) = 2. část (Status)
    names_pattern = "(.*) average FPS (.*) \\(FPS\\)", 
    values_to = "FPS" # Kam uložit hodnoty
  ) %>%
  drop_na(FPS)
df_long
```

```r
summary_full <- df_long %>%
  group_by(Karta, Status) %>%
  summarize(
    pocet = n(),
    prumer = mean(FPS),
    median = median(FPS),
    minimum = min(FPS),
    maximum = max(FPS),
    rozsah = maximum - minimum,
    dolni_kvartil = quantile(FPS, 0.25),
    horni_kvartil = quantile(FPS, 0.75),
    iqr = IQR(FPS), # Mezokvartilové rozpětí
    .groups = 'drop' # Zruší seskupení
  )

summary_full
```

```r
karty_k_analyze <- c("Nvidia RTX 3070 Ti", "AMD Radeon RX 6800 XT")

df_rozdily <- df_long %>%
  # 1. Krok: Vybereme jen relevantní karty
  filter(Karta %in% karty_k_analyze) %>%
  
  # 2. Krok: Přetočíme data, aby 'release' a 'patched' byly sloupce
  pivot_wider(
    names_from = Status,  # Sloupce se budou jmenovat 'release' a 'patched'
    values_from = FPS     # Hodnoty se vezmou z FPS
  ) %>%
  
  # 3. Krok: Vytvoříme nový sloupec s rozdílem
  mutate(
    FPS_narust = patched - release
  )

# Podívej se na výsledek!
print(df_rozdily)
```

```r
summary_narust <- df_rozdily %>%
  group_by(Karta) %>%
  summarize(
    pocet = n(),
    prumerny_narust = mean(FPS_narust),
    median_narustu = median(FPS_narust),
    min_narust = min(FPS_narust),
    max_narust = max(FPS_narust),
    rozsah_narustu = max_narust - min_narust,
    Q1_narustu = quantile(FPS_narust, 0.25),
    Q3_narustu = quantile(FPS_narust, 0.75),
    iqr_narustu = IQR(FPS_narust)
  )

summary_narust
```

```r
# 2. --- POMOCNÉ FUNKCE DLE PDF ---
# Tato funkce určí počet platných cifer pro 's' na základě velikosti výběru 'n'
# [Zdroj: zaokrouhlovani.pdf, Tabulka 1, str. 3] 
get_platne_cifry <- function(n) {
  if (n <= 10) {
    return(1)
  } else if (n <= 30) {
    return(2)
  } else if (n <= 2000) {
    return(3)
  } else {
    return(4)
  }
}

# Tato funkce zaokrouhlí směrodatnou odchylku NAHORU na daný počet platných cifer
# [Zdroj: zaokrouhlovani.pdf, Sekce 3.1]
zaokrouhli_s <- function(s_val, n_val) {
  
  # --- NOVÁ KONTROLA ---
  # Ošetření okrajových případů:
  # 1. Pokud je s_val NA nebo NaN (Not a Number, např. z sd(c(1))), vrať NA
  if (is.na(s_val) || is.nan(s_val)) {
    return(NA)
  }
  # 2. Pokud je s_val 0 (všechny hodnoty jsou stejné), vrať 0
  if (s_val == 0) {
    return(0)
  }
  # --- KONEC NOVÉ KONTROLY ---
  
  # Získáme počet platných cifer dle velikosti výběru
  platne_cifry <- get_platne_cifry(n_val)
  
  # Matematický postup pro zaokrouhlení nahoru na platné cifry
  # Tento kód se teď spustí jen pokud je s_val kladné číslo
  magnituda <- 10^floor(log10(abs(s_val)))
  s_adj <- (s_val / magnituda) * (10^(platne_cifry - 1))
  s_ceil <- ceiling(s_adj) # Zaokrouhlení nahoru
  s_final <- (s_ceil / (10^(platne_cifry - 1))) * magnituda
  
  return(s_final)
}

# Tato funkce zjistí počet desetinných míst (přesnost) ze zaokrouhleného 's'
# [Zdroj: zaokrouhlovani.pdf, Sekce 3.2] 
get_digits <- function(rounded_val) {
  if (is.na(rounded_val)) {
    return(0) 
  }
  val_char <- as.character(rounded_val)
  if (!grepl("\\.", val_char)) {
    return(0) # Je to celé číslo
  } else {
    # Spočítá počet znaků za desetinnou čárkou
    return(nchar(strsplit(val_char, "\\.")[[1]][2]))
  }
}
# --- KONEC POMOCNÝCH FUNKCÍ ---


# 3. VÝPOČET STATISTIK (Předpokládáme, že 'df_rozdily' existuje)

# Nejprve spočítáme VŠECHNY statistiky v "surovém" stavu 
raw_stats <- df_rozdily %>%
  group_by(Karta) %>%
  summarize(
    pocet = n(),
    prumer_raw = mean(FPS_narust, na.rm = TRUE),
    median_raw = median(FPS_narust, na.rm = TRUE),
    q1_raw = quantile(FPS_narust, 0.25, na.rm = TRUE),
    q3_raw = quantile(FPS_narust, 0.75, na.rm = TRUE),
    min_raw = min(FPS_narust, na.rm = TRUE),
    max_raw = max(FPS_narust, na.rm = TRUE),
    s_raw = sd(FPS_narust, na.rm = TRUE),
    # Použijeme balíček 'moments'
    sikmost_raw = skewness(FPS_narust, na.rm = TRUE),
    # PDF zmiňuje, že 'moments' vrací špičatost + 3[cite: 55].
    # Pro porovnání s normálním rozdělením (kde je špičatost 0) musíme 3 odečíst.
    spicatost_raw = kurtosis(FPS_narust, na.rm = TRUE) - 3 
  )

# 4. APLIKACE ZAOKROUHLOVACÍCH PRAVIDEL
final_summary <- raw_stats %>%
  rowwise() %>% # Nutné, aby se funkce aplikovaly na každý řádek (kartu) zvlášť
  mutate(
    # 1. Nejdřív zaokrouhlit 's' dle pravidel [cite: 33, 37]
    smerodatna_odchylka = zaokrouhli_s(s_raw, pocet),
    
    # 2. Zjistíme přesnost (počet des. míst) zaokrouhleného 's' 
    presnost_s = get_digits(smerodatna_odchylka),
    
    # 3. Zaokrouhlíme míry polohy dle 'presnost_s' 
    rozsah = round(max_raw - min_raw, digits = presnost_s),
    minimum = round(min_raw, digits = presnost_s),
    dolni_kvartil = round(q1_raw, digits = presnost_s),
    median = round(median_raw, digits = presnost_s),
    prumer = round(prumer_raw, digits = presnost_s),
    horni_kvartil = round(q3_raw, digits = presnost_s),
    maximum = round(max_raw, digits = presnost_s),
    
    # 4. Zaokrouhlíme šikmost a špičatost na desetiny 
    sikmost = round(sikmost_raw, digits = 1),
    spicatost = round(spicatost_raw, digits = 1)
  ) %>%
  ungroup() # Zrušíme rowwise

# 5. FINÁLNÍ VÝBĚR A SEŘAZENÍ DLE TVÉHO ZADÁNÍ
final_output <- final_summary %>%
  select(
    Karta,
    rozsah,
    minimum,
    dolni_kvartil,
    median,
    prumer,
    horni_kvartil,
    maximum,
    smerodatna_odchylka,
    sikmost,
    spicatost,
    pocet # Ponecháme 'pocet' pro kontrolu, jak se zaokrouhlilo 's'
  )

# Zobrazení finální tabulky
final_output
```

```r
# Předpokládáme, že 'raw_stats_cleaned' a 'final_summary_cleaned' existují

# 1. Získáme surové hodnoty a přesnost
surove_hodnoty <- raw_stats_cleaned %>%
  select(Karta, prumer_raw, s_raw)
  
presnost_data <- final_summary_cleaned %>%
  select(Karta, presnost_s)

# 2. Spojíme data pro finální výpočet
data_pro_intervaly <- surove_hodnoty %>%
  left_join(presnost_data, by = "Karta")

# 3. Vypočítáme intervaly s NOVÝM zaokrouhlováním
#    (floor pro dolní, ceiling pro horní)
intervaly_final <- data_pro_intervaly %>%
  rowwise() %>%
  mutate(
    # --- Výpočet 2-sigma (pro 95 %) ---
    lower_2s_raw = prumer_raw - (2 * s_raw),
    upper_2s_raw = prumer_raw + (2 * s_raw),
    
    # floor(x * 10^p) / 10^p  == zaokrouhlení dolů na 'p' míst
    # ceiling(x * 10^p) / 10^p == zaokrouhlení nahoru na 'p' míst
    p = 10^presnost_s, # Násobič pro posun desetinné čárky
    
    interval_2s_dolni = floor(lower_2s_raw * p) / p,
    interval_2s_horni = ceiling(upper_2s_raw * p) / p,

    # --- Výpočet 3-sigma (pro 99.7 %) ---
    lower_3s_raw = prumer_raw - (3 * s_raw),
    upper_3s_raw = prumer_raw + (3 * s_raw),

    interval_3s_dolni = floor(lower_3s_raw * p) / p,
    interval_3s_horni = ceiling(upper_3s_raw * p) / p
  ) %>%
  ungroup() %>%
  select(
    Karta,
    interval_2s_dolni, interval_2s_horni,
    interval_3s_dolni, interval_3s_horni
  )

# 4. Vytiskneme výslednou tabulku
print("--- Intervaly (2s a 3s) se zaokrouhlením DLE SKRIPT ---")
intervaly_final
```

```r
# Předpokládáme, že existuje 'df_rozdily_cleaned' a 'final_summary_cleaned' (pro přesnost)

# 1. Shapiro-Wilkův test normality
# H0: Data mají normální rozdělení
# H1: Data nemají normální rozdělení
shapiro_vysledky <- df_rozdily_cleaned %>%
  group_by(Karta) %>%
  shapiro_test(FPS_narust)

print("--- 1. SHAPIRO-WILKŮV TEST (Ověření normality) ---")
print(shapiro_vysledky)


# 2. T-test a Intervalové odhady
# Testujeme, zda je nárůst statisticky významně odlišný od 0
ttest_results <- df_rozdily_cleaned %>%
  group_by(Karta) %>%
  summarise(
    # Spočítáme t-test pro každou skupinu
    t_test_obj = list(t.test(FPS_narust, mu = 0, conf.level = 0.95))
  ) %>%
  rowwise() %>%
  mutate(
    p_value = t_test_obj$p.value,
    # Vytažení surových mezí intervalu spolehlivosti
    ci_lower_raw = t_test_obj$conf.int[1],
    ci_upper_raw = t_test_obj$conf.int[2],
    estimate_mean = t_test_obj$estimate
  ) %>%
  # Připojíme info o přesnosti zaokrouhlování (počet des. míst) z minula
  left_join(final_summary_cleaned %>% select(Karta, presnost_s), by = "Karta") %>%
  
  # APLIKACE ZAOKROUHLOVACÍCH PRAVIDEL PRO INTERVALOVÉ ODHADY
  # Dolní mez -> DOLŮ, Horní mez -> NAHORU 
  mutate(
    p = 10^presnost_s, # Pomocná proměnná pro posun desetinné čárky
    
    # Zaokrouhlení intervalu spolehlivosti
    CI_dolni = floor(ci_lower_raw * p) / p,
    CI_horni = ceiling(ci_upper_raw * p) / p,
    
    # Zaokrouhlení bodového odhadu (průměru) klasicky
    Bodovy_odhad = round(estimate_mean, digits = presnost_s),
    
    # Formátování p-hodnoty na tisíciny [cite: 87]
    p_hodnota_text = sprintf("%.3f", p_value)
  ) %>%
  select(Karta, p_hodnota_text, Bodovy_odhad, CI_dolni, CI_horni)

print("--- 2. VÝSLEDKY TESTŮ A INTERVALY SPOLEHLIVOSTI ---")
print(ttest_results)
```

```r

# Předpokládáme, že 'df_rozdily_cleaned' je tvůj vyčištěný dataset

floor_dec <- function(x, level=1) floor(x * 10^level) / 10^level
ceiling_dec <- function(x, level=1) ceiling(x * 10^level) / 10^level

# --- 1. TABULKA: OBOUSTRANNÝ TEST (95% Intervalový odhad) ---
# Testujeme odlišnost od nuly na obě strany
tabulka_oboustranna <- df_rozdily_cleaned %>%
  group_by(Karta) %>%
  summarise(
    test = list(t.test(FPS_narust, mu = 0, alternative = "two.sided", conf.level = 0.95))
  ) %>%
  rowwise() %>%
  mutate(
    Typ_testu = "Oboustranný (Two-sided)",
    # Interval
    Dolni_mez = sprintf("%.3f", floor_dec(test$conf.int[1], 3)),
    Horni_mez = sprintf("%.3f", floor_dec(test$conf.int[2], 3)),
    # P-hodnota
    P_value = sprintf("%.3f", ceiling_dec(test$p.value, 3)) # nebo %.5f pro větší detail
  ) %>%
  select(Karta, Typ_testu, Dolni_mez, Horni_mez, P_value)

print("--- 1. OBOUSTRANNÝ TEST (Klasický interval) ---")
print(tabulka_oboustranna)


# --- 2. TABULKA: PRAVOSTRANNÝ TEST (Hledáme Dolní mez) ---
# Testujeme, zda je nárůst VĚTŠÍ než 0.
# Interval bude ve tvaru: ( Dolní_mez : Inf )
tabulka_pravostranna <- df_rozdily_cleaned %>%
  group_by(Karta) %>%
  summarise(
    test = list(t.test(FPS_narust, mu = 0, alternative = "greater", conf.level = 0.95))
  ) %>%
  rowwise() %>%
  mutate(
    Typ_testu = "Pravostranný (Greater)",
    # Dolní mez je číslo, Horní je nekonečno
    Dolni_mez = sprintf("%.3f", floor_dec(test$conf.int[1], 3)),
    Horni_mez = "Inf",
    # P-hodnota
    P_value = sprintf("%.3f", ceiling_dec(test$p.value, 3))
  ) %>%
  select(Karta, Typ_testu, Dolni_mez, Horni_mez, P_value)

print("--- 2. PRAVOSTRANNÝ TEST (Dolní odhad / Levostranný interval ve formuláři) ---")
print(tabulka_pravostranna)


# --- 3. TABULKA: LEVOSTRANNÝ TEST (Hledáme Horní mez) ---
# Testujeme, zda je nárůst MENŠÍ než 0 (což tady asi zamítneme, ale pro úplnost).
# Interval bude ve tvaru: ( -Inf : Horní_mez )
tabulka_levostranna <- df_rozdily_cleaned %>%
  group_by(Karta) %>%
  summarise(
    test = list(t.test(FPS_narust, mu = 0, alternative = "less", conf.level = 0.95))
  ) %>%
  rowwise() %>%
  mutate(
    Typ_testu = "Levostranný (Less)",
    # Dolní mez je mínus nekonečno, Horní je číslo
    Dolni_mez = "-Inf",
    Horni_mez = sprintf("%.3f", floor_dec(test$conf.int[2], 3)),
    # P-hodnota
    P_value = sprintf("%.3f", ceiling_dec(test$p.value, 3))
  ) %>%
  select(Karta, Typ_testu, Dolni_mez, Horni_mez, P_value)

print("--- 3. LEVOSTRANNÝ TEST (Horní odhad) ---")
print(tabulka_levostranna)
```

```r
library(tidyverse)

# 1. Příprava dat (rozdělíme na dva vektory)
data_nvidia <- df_rozdily_cleaned %>% 
  filter(Karta == "Nvidia RTX 3070 Ti") %>% 
  pull(FPS_narust)

data_amd <- df_rozdily_cleaned %>% 
  filter(Karta == "AMD Radeon RX 6800 XT") %>% 
  pull(FPS_narust)

# 2. Funkce pro výpočet všech intervalů
spocitej_intervaly <- function(x, y, conf.level = 0.95) {
  
  # A. OBOUSTRANNÝ (Two-sided) -> Klasický interval (od - do)
  t2 <- t.test(x, y, alternative = "two.sided", var.equal = TRUE, conf.level = conf.level)
  
  # B. PRAVOSTRANNÝ (Greater) -> Hledáme DOLNÍ MEZ ( X : Inf )
  # Testujeme, že rozdíl > 0 (že Nvidia je lepší)
  tr <- t.test(x, y, alternative = "greater", var.equal = TRUE, conf.level = conf.level)
  
  # C. LEVOSTRANNÝ (Less) -> Hledáme HORNÍ MEZ ( -Inf : Y )
  tl <- t.test(x, y, alternative = "less", var.equal = TRUE, conf.level = conf.level)
  
  # Tabulka výsledků
  tibble(
    Typ_testu = c("Oboustranný", "Pravostranný (Dolní mez)", "Levostranný (Horní mez)"),
    
    # Surové hodnoty
    Dolni_raw = c(t2$conf.int[1], tr$conf.int[1], tl$conf.int[1]),
    Horni_raw = c(t2$conf.int[2], tr$conf.int[2], tl$conf.int[2]),
    P_value = c(t2$p.value, tr$p.value, tl$p.value)
  ) %>%
    rowwise() %>%
    mutate(
      # --- ZAOKROUHLOVÁNÍ (Dolní dolů, Horní nahoru) ---
      
      # Verze na 3 desetinná místa (pro jistotu)
      Dolni_3des = ifelse(Dolni_raw == -Inf, "-Inf", sprintf("%.3f", ceiling_dec(Dolni_raw, 3))),
      Horni_3des = ifelse(Horni_raw == Inf, "Inf", sprintf("%.3f", ceiling_dec(Horni_raw, 3))),
      
      # Verze na 1 desetinné místo (podle skript pro n=6)
      Dolni_1des = ifelse(Dolni_raw == -Inf, "-Inf", sprintf("%.1f", floor_dec(Dolni_raw, 1))),
      Horni_1des = ifelse(Horni_raw == Inf, "Inf", sprintf("%.1f", ceiling_dec(Horni_raw, 1))),
      
      P_val_txt = sprintf("%.3f", ceiling_dec(P_value, 3))
    ) %>%
    select(Typ_testu, Dolni_3des, Horni_3des, Dolni_1des, Horni_1des, P_val_txt)
}

# 3. Spuštění a výpis
vysledky_rozdil <- spocitej_intervaly(data_nvidia, data_amd)

print("--- INTERVALY PRO ROZDÍL (Nvidia - AMD) ---")
print(vysledky_rozdil)
```

```r
# --- HELPER FUNCTIONS ---
get_platne_cifry <- function(n) {
  if (n <= 10) return(1)
  if (n <= 30) return(2)
  if (n <= 2000) return(3)
  return(4)
}

zaokrouhli_s <- function(s_val, n_val) {
  if (is.na(s_val) || is.nan(s_val)) return(NA)
  if (s_val == 0) return(0)
  platne_cifry <- get_platne_cifry(n_val)
  magnituda <- 10^floor(log10(abs(s_val)))
  s_adj <- (s_val / magnituda) * (10^(platne_cifry - 1))
  s_ceil <- ceiling(s_adj)
  s_final <- (s_ceil / (10^(platne_cifry - 1))) * magnituda
  return(s_final)
}

get_digits <- function(rounded_val) {
  if (is.na(rounded_val)) return(0)
  val_char <- as.character(rounded_val)
  if (!grepl("\\.", val_char)) return(0)
  return(nchar(strsplit(val_char, "\\.")[[1]][2]))
}

# --- 1. LOAD & PREPARE DATA ---
raw_data <- read_excel("FPS_data.xlsx", sheet = "Vysledky mereni")

# Convert to long format and calculate increase
df_diffs_all <- raw_data %>%
  pivot_longer(
    cols = -id,
    names_to = c("Karta", "Status"),
    names_pattern = "(.*) average FPS (.*) \\(FPS\\)",
    values_to = "FPS"
  ) %>%
  drop_na(FPS) %>%
  pivot_wider(names_from = Status, values_from = FPS) %>%
  mutate(FPS_narust = patched - release)

# --- 2. STATS WITH OUTLIERS ---
stats_with_outliers <- df_diffs_all %>%
  group_by(Karta) %>%
  summarize(
    n = n(),
    mean_raw = mean(FPS_narust),
    median_raw = median(FPS_narust),
    s_raw = sd(FPS_narust),
    skew = skewness(FPS_narust),
    kurt = kurtosis(FPS_narust) - 3
  ) %>%
  rowwise() %>%
  mutate(
    s_rounded = zaokrouhli_s(s_raw, n),
    precision = get_digits(s_rounded),
    Mean = round(mean_raw, precision),
    Median = round(median_raw, precision),
    SD = s_rounded
  ) %>%
  select(Karta, n, Mean, Median, SD, skew, kurt)

print("=== STATS WITH OUTLIERS ===")
print(stats_with_outliers)

# --- 3. REMOVE OUTLIERS ---
outliers <- df_diffs_all %>% group_by(Karta) %>% identify_outliers(FPS_narust)
df_clean_all <- df_diffs_all %>% anti_join(outliers, by = "id")

# --- 4. STATS WITHOUT OUTLIERS ---
stats_clean <- df_clean_all %>%
  group_by(Karta) %>%
  summarize(
    n = n(),
    mean_raw = mean(FPS_narust),
    median_raw = median(FPS_narust),
    s_raw = sd(FPS_narust),
    skew = skewness(FPS_narust),
    kurt = kurtosis(FPS_narust) - 3
  ) %>%
  rowwise() %>%
  mutate(
    s_rounded = zaokrouhli_s(s_raw, n),
    precision = get_digits(s_rounded),
    Mean = round(mean_raw, precision),
    Median = round(median_raw, precision),
    SD = s_rounded
  ) %>%
  select(Karta, n, Mean, Median, SD, skew, kurt)

print("=== STATS WITHOUT OUTLIERS ===")
print(stats_clean)

# --- 5. STATISTICAL TESTS (CLEAN DATA) ---

# Shapiro-Wilk (Normality)
shapiro_res <- df_clean_all %>% group_by(Karta) %>% shapiro_test(FPS_narust)
print("=== SHAPIRO-WILK TEST ===")
print(shapiro_res)

# Bartlett's Test (Homogeneity of Variances)
bartlett_res <- bartlett.test(FPS_narust ~ Karta, data = df_clean_all)
print("=== BARTLETT'S TEST ===")
print(bartlett_res)
print(paste("P-value:", sprintf("%.3f", bartlett_res$p.value)))

# Levene's Test (Robust Homogeneity)
levene_res <- leveneTest(FPS_narust ~ Karta, data = df_clean_all)
print("=== LEVENE'S TEST ===")
print(paste("P-value:", sprintf("%.3f", levene_res$`Pr(>F)`[1])))
```

```r
library(tidyverse)
library(readxl)
library(rstatix)

# --- 1. NAČTENÍ A PŘÍPRAVA DAT ---
raw_data <- read_excel("FPS_data.xlsx", sheet = "Vysledky mereni")

df_all <- raw_data %>%
  pivot_longer(
    cols = -id,
    names_to = c("Karta", "Status"),
    names_pattern = "(.*) average FPS (.*) \\(FPS\\)",
    values_to = "FPS"
  ) %>%
  drop_na(FPS) %>%
  pivot_wider(names_from = Status, values_from = FPS) %>%
  mutate(FPS_narust = patched - release)

# --- 2. CLEAN DATA ---
outliers <- df_all %>% 
  group_by(Karta) %>% 
  identify_outliers(FPS_narust)

df_clean <- df_all %>% 
  anti_join(outliers, by = "id")

# Barvy (modrá paleta)
barvy <- colorRampPalette(c("lightblue", "dodgerblue3"))(4)

# ==============================================================================
# 1. BOXPLOT - PŮVODNÍ DATA
# ==============================================================================
svg("boxplot_puvodni.svg", width = 10, height = 8)
par(mar = c(12, 5, 4, 2) + 0.1) 

boxplot(
  FPS_narust ~ Karta,
  data = df_all,
  # Učitel nechce obecné názvy, specifikujeme proměnnou
  main = "Rozdělení nárůstu FPS pro jednotlivé karty (Původní data)",
  ylab = "Nárůst FPS",
  xlab = "",
  col = barvy,
  border = "black",
  las = 2
)
mtext("Grafická karta", side = 1, line = 9)
grid()
dev.off()

# ==============================================================================
# 2. BOXPLOT - BEZ OUTLIERŮ
# ==============================================================================
svg("boxplot_bez_outlieru.svg", width = 10, height = 8)
par(mar = c(12, 5, 4, 2) + 0.1)

boxplot(
  FPS_narust ~ Karta,
  data = df_clean,
  main = "Rozdělení nárůstu FPS pro jednotlivé karty (Bez odlehlých pozorování)",
  ylab = "Nárůst FPS",
  xlab = "",
  col = barvy,
  border = "black",
  las = 2
)
mtext("Grafická karta", side = 1, line = 9)
grid()
dev.off()

# ==============================================================================
# 3. HISTOGRAMY (Sjednocená osa X + Oprava názvů)
# ==============================================================================
svg("histogramy_souhrn.svg", width = 12, height = 10)
par(mar = c(5, 4, 4, 2) + 0.1)
par(mfrow = c(2, 2)) 

karty <- unique(df_clean$Karta)

# --- VÝPOČET GLOBÁLNÍCH LIMITŮ PRO OSU X ---
# Aby byly grafy srovnatelné, musí mít stejný rozsah osy X
global_min_x <- min(df_clean$FPS_narust)
global_max_x <- max(df_clean$FPS_narust)
my_xlim <- c(floor(global_min_x), ceiling(global_max_x)) # Zaokrouhlíme pro hezčí osy

for (i in 1:length(karty)) {
  karta_nazev <- karty[i]
  data_subset <- df_clean$FPS_narust[df_clean$Karta == karta_nazev]
  
  # Výpočet limitů osy Y (jako minule)
  dens <- density(data_subset, na.rm = TRUE)
  h_info <- hist(data_subset, plot = FALSE)
  max_y <- max(c(h_info$density, dens$y))
  my_ylim <- c(0, max_y * 1.1)
  
  hist(
    data_subset,
    freq = FALSE,
    # POUZE NÁZEV KARTY (bez slova Histogram)
    main = karta_nazev, 
    xlab = "Nárůst FPS",
    ylab = "Hustota pravděpodobnosti", # Oprava dle učitele
    col = barvy[i],
    border = "black",
    ylim = my_ylim,
    xlim = my_xlim # Aplikace sjednocené osy X
  )
  lines(dens, col = "red", lwd = 2)
}
par(mfrow = c(1, 1))
dev.off()

# ==============================================================================
# 4. Q-Q GRAFY (Oprava názvů)
# ==============================================================================
svg("qq_ploty_souhrn.svg", width = 12, height = 10)
par(mfrow = c(2, 2))

for (i in 1:length(karty)) {
  karta_nazev <- karty[i]
  data_subset <- df_clean$FPS_narust[df_clean$Karta == karta_nazev]
  
  qqnorm(
    data_subset,
    # POUZE NÁZEV KARTY (bez slova Q-Q Plot)
    main = karta_nazev,
    xlab = "Teoretické kvantily",
    ylab = "Výběrové kvantily",
    pch = 19,
    col = "darkblue"
  )
  qqline(data_subset, col = "red", lwd = 2)
  grid()
}
par(mfrow = c(1, 1))
dev.off()
```
