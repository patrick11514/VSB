```r
library(tidyverse)

# 1. Načtení dat (uprav si název souboru)
data_pf <- readxl::read_excel("portfolia_vynosy.xlsx") 

# --- ŘEŠENÍ PRO STRATEGII A ---

# A) Příprava čísel
# Vyfiltrujeme jen sloupec pro strategii A (předpokládám název 'Vynos_A' nebo podobně)
# Pokud máš data pod sebou, použij filter(Strategie == "A")
vynosy_A <- data_pf$Vynos_A # <--- ZKONTROLUJ NÁZEV SLOUPCE!

# Počet všech pokusů (n)
n_A <- sum(!is.na(vynosy_A)) 

# Počet úspěchů (x) -> Kde je výnos > 0
x_A <- sum(vynosy_A > 0, na.rm = TRUE)

print(paste("Strategie A: Úspěchů", x_A, "z", n_A))

# B) Vlastní test
# x = počet úspěchů, n = počet celkem, p = 0.7 (testovaná hodnota)
# alternative = "greater" (protože nás zajímá, jestli PŘEVYŠUJE)
test_A <- prop.test(x = x_A, n = n_A, p = 0.7, alternative = "greater", conf.level = 0.90)

print("--- VÝSLEDEK TESTU STRATEGIE A ---")
print(test_A)
print(paste("P-hodnota:", format.pval(test_A$p.value, digits = 4)))


# --- ŘEŠENÍ PRO STRATEGII B ---
# (To samé, jen pro béčko)
vynosy_B <- data_pf$Vynos_B # <--- ZKONTROLUJ NÁZEV SLOUPCE!

n_B <- sum(!is.na(vynosy_B))
x_B <- sum(vynosy_B > 0, na.rm = TRUE)

print(paste("Strategie B: Úspěchů", x_B, "z", n_B))

test_B <- prop.test(x = x_B, n = n_B, p = 0.7, alternative = "greater", conf.level = 0.90)

print("--- VÝSLEDEK TESTU STRATEGIE B ---")
print(test_B)
print(paste("P-hodnota:", format.pval(test_B$p.value, digits = 4)))
```

```r
# ==============================================================================
# ✅ KONTROLA PŘEDPOKLADU (Minimální n) - TAHÁK
# ==============================================================================
# Vzorec: n > 9 / (p * (1-p))
# ------------------------------------------------------------------------------

# 1. ZADEJ HODNOTY (Buď ručně, nebo použij proměnné z minula p_A, n_A)
p_test <- p_A  # Zde dosaď bodový odhad (např. 0.7654)
n_test <- n_A  # Zde dosaď skutečný počet řádků (např. 80)

# 2. VÝPOČET
vypocet <- 9 / (p_test * (1 - p_test))
min_n   <- ceiling(vypocet) # DŮLEŽITÉ: Zaokrouhlujeme vždy NAHORU!

# 3. VÝPIS PRO ODPOVĚĎ
print(paste("Bodový odhad p:", round(p_test, 4)))
print(paste("Vypočtené minimum (Odpověď do testu):", min_n))
print(paste("Máme k dispozici n:", n_test))

if(n_test > min_n) {
  print("✅ PŘEDPOKLAD JE SPLNĚN (Máme dost dat).")
} else {
  print("❌ PŘEDPOKLAD NENÍ SPLNĚN (Máme málo dat).")
}
```
