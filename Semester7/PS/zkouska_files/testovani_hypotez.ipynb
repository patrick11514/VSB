```python
# ==============================================================================
# üß© TEST NEZ√ÅVISLOSTI (Ch√≠-kvadr√°t / Chi-Square) - TAH√ÅK
# ==============================================================================
# Kdy pou≈æ√≠t: "Z√°vis√≠ barva oƒç√≠ na pohlav√≠?", "Souvis√≠ strategie se ziskem?"
# Data: Pot≈ôebuje≈° dva sloupce s textem/kategoriemi (nebo ƒç√≠sla p≈ôeveden√° na kategorie).
# ==============================================================================

library(tidyverse)

# 1. P≈ò√çPRAVA DAT (Vytvo≈ôen√≠ kategori√≠, pokud nejsou)
# P≈ô√≠klad: P≈ôevod ƒç√≠sel na "Zisk"/"Ztr√°ta"
data_prep <- data %>%
  mutate(Kat1 = Strategie,                                            # Prvn√≠ kategorie
         Kat2 = ifelse(Rocni_Vynosnost_v_Procentech > 0, "ANO", "NE")) # Druh√° kategorie

# 2. VYTVO≈òEN√ç TABULKY
tabulka_test <- table(data_prep$Kat1, data_prep$Kat2)
print(tabulka_test)

# 3. TEST
test_vysledek <- chisq.test(tabulka_test, correct = FALSE)
print(test_vysledek)

# 4. VYHODNOCEN√ç
print(paste("P-hodnota:", format.pval(test_vysledek$p.value, digits = 5)))

if (test_vysledek$p.value < 0.05) {
  print("‚úÖ ZAM√çT√ÅME H0 -> Existuje Z√ÅVISLOST.")
} else {
  print("‚ùå NEZAM√çT√ÅME H0 -> Veliƒçiny jsou NEZ√ÅVISL√â.")
}
```
