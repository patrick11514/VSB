```r
A = rnorm(30, 100, 10)
B = rnorm(33, 100, 10)
```

```r
boxplot(A)
boxplot(B)
```

```r
kruskal.test(A ~ B)
```

```r
A.df = as.data.frame(A)
B.df = as.data.frame(B)
A.df
B.df
```

```r
A.df$typ = "A"
B.df$typ = "B"
colnames(A.df) = c("hodnoty", "typ")
colnames(B.df) = c("hodnoty", "typ")
head(A.df)
head(B.df)
```

```r
data.s = rbind(A.df,B.df)
boxplot(data.s$hodnoty ~ data.s$typ)
```
